# Does N$_{2}$ _really_ behave like an Ideal Gas?

Lets compute the compressibility factor $Z$ for N$_{2}$ for a range of temperatures $T$ using data from NIST 

```julia
# Load Julia packages -
using DataFrames
using CSV
using PyPlot

# load my code -
include("./code/Compressibility.jl")

# Load the N2 data -
nitrogen_data_table_198K = CSV.read("Nitrogen-NIST-198K.txt",delim='\t');
nitrogen_data_table_298K = CSV.read("Nitrogen-NIST-298K.txt",delim='\t');
nitrogen_data_table_398K = CSV.read("Nitrogen-NIST-398K.txt",delim='\t');
nitrogen_data_table_900K = CSV.read("Nitrogen-NIST-900K.txt",delim='\t');
```

```julia
nitrogen_data_table_900K
```

```julia
# call my code to compute the commpressibility -
(Z_900,P_900) = calculateCompressibility(nitrogen_data_table_900K);
(Z_298,P_298) = calculateCompressibility(nitrogen_data_table_298K);
(Z_398,P_398) = calculateCompressibility(nitrogen_data_table_398K);
(Z_198,P_198) = calculateCompressibility(nitrogen_data_table_198K);


# plot -
index_upper = 100;
plot(P_900[1:index_upper],Z_900[1:index_upper],"r",label="T = 900K");
plot(P_298[1:index_upper],Z_298[1:index_upper],"b",label="T = 298K");
plot(P_398[1:index_upper],Z_398[1:index_upper],"g",label="T = 398K");
plot(P_198[1:index_upper],Z_198[1:index_upper],"k",label="T = 198K");
axis([0,100,0.8,2]);

# axis labels -
xlabel("Pressure (MPa)",fontsize=16);
ylabel(L"Compressibility $Z$",fontsize=16);

# legend -
legend(loc="upper left");
```
