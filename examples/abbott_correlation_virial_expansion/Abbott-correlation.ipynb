### Example: Compressibility factor $Z$ model using the Abbott correlations for Cyclohexane ###

#### Download data from NIST and compute the Z factor experimentally. 

```julia
# includes -
using PyPlot
using DataFrames
using CSV

# load my code -
include("./code/Compressibility.jl")

# Critical parameters -
Tc_cyclohexane = 554.0 # units: K
Pc_cyclohexane = 4.07 # MPa
Tc_methannol = 513.0 # units: K
Pc_methanol = 8.1 # MPa

# Load the Cyclohexane data -
methanol_data_table = CSV.read("Methanol-NIST-615K.txt",delim='\t');
cyclohexane_data_table = CSV.read("Cyclohexane-NIST-664K.txt",delim='\t');

# === Tr = 1.2 ===========================================================
# Compute the compressibility -
(Z_cyclohexane,P_cyclohexane) = calculateCompressibility(cyclohexane_data_table)
(Z_methanol,P_methanol) = calculateCompressibility(methanol_data_table)

# Compute the reduced pressure -
P_cyclohexane_r = (P_cyclohexane).*(1/Pc_cyclohexane);
P_methanol_r = (P_methanol).*(1/Pc_methanol);
# ========================================================================
```

#### Compute the Z factor for Cyclohexane using the Abbott correlation models 

```julia
# call the Abbott model code -
T_cyclohexane = 664.0 # K
T_methanol = 615.0 # K
w_Cyclohexane = 0.212 # R. C. Reid, J. M. Prausnitz, and B. E. Poling, 1987, The Properties of Gases and Liquids, 4th Ed. New York: McGraw-Hill.
w_Methanol = 0.556 # R. C. Reid, J. M. Prausnitz, and B. E. Poling, 1987, The Properties of Gases and Liquids, 4th Ed. New York:
(Z_array_AC,Z0_array_AC,Z1_array_AC,P) = compute_abbott_compressibility(P_cyclohexane,T_cyclohexane,Pc_cyclohexane,Tc_cyclohexane,w_Cyclohexane);
(Z_array_AM,Z0_array_AM,Z1_array_AM,P) = compute_abbott_compressibility(P_methanol,T_methanol,Pc_methanol,Tc_methannol,w_Methanol);
```

```julia
# plot -
#plot(P_cyclohexane_r,Z_array_AC,"b--",label="Cyclohexane Tr = 1.2 Abbott")
#plot(P_cyclohexane_r,Z_cyclohexane,"b",label="Cyclohexane Tr = 1.2");
plot(P_methanol_r,Z_array_AM,"r--",label="Methanol Tr = 1.2 Abbott")
plot(P_methanol_r,Z_methanol,"r",label="Methanol Tr = 1.2");

# label the axis -
xlabel(L"Reduced pressure P$_{r}$",fontsize=16);
ylabel(L"Compressibility factor $Z$",fontsize=16);
legend(loc="upper right");
```
